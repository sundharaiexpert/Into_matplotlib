```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
print(car_sales["Price"][3])
```

```python
# Remove last two zeros 
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Toral Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Drop the typo error column from DataFrame
car_sales.drop(columns="Toral Sales", axis=1, inplace=True)
car_sales
```

```python
# Plot the Total sales
car_sales.plot(x="Sale Date", y="Total Sales");
```

## Which on should you use ? (pyplot vs matplotlib OO method)

* when plotting quickly use pyplot method
* when use advanced then use OO method

# Plotting dataframe 6


```python
over_50.target.values
```

```python
# OO method from scratch 
fig , ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

# Add alegend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line in axis

ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Plotting from Dataframe 7
# add subplots 

# Subplots of chol, age. thatlach 

fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)
# Add data to ax0
scatter = ax0.scatter(x =over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# custpmize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
      ylabel="Cholesterol")

# Add alegend ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a mean line in axis
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x =over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# custpmize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
      xlabel="Age",
      ylabel="MAx Heart Rate")
# Add alegend ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# Add a mean line in axis
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle="--");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
