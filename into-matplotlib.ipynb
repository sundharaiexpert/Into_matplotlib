## Matplotlib exmaple workflow


```python
%matplotlib inline
import matplotlib.pyplot as plt

#1.prepare data
x = [1,2,3,4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) #Widht, height

# 3.plot the data
ax.plot(x, y)

# 4.customizxe
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5.Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")


```

# Making figures with Numpy arrays
We are making 
* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* subplot

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:100]
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# using the same data for Scatter Plot(lots of dotts)
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictinary
nut_butter_prices = {"Almod butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter" : 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),height= nut_butter_prices.values() );
ax.set(title="Butter prices in store",
      ylabel="Price ($)");
```

```python
# Histogram and Subplots
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

```python
# Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
#Two options for subplots - 1 Option
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```
